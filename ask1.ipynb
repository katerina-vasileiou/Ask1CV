```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt


inputPicName = 'trikoupi6.png'      
outputPicName = 'tresholdedPic1'               
thdKey = 110                      

img = Image.open(inputPicName)  

pixelsData_array = np.array(img)            
width, height = img.size                    

# check if the opened image is grayscale mode
isGray = True                               
imgNew = img.convert("RGB")                 

for i in range(width):                      
    for j in range(height):
        r, g, b = imgNew.getpixel((i, j))

        if r != g != b:
            isGray = False
            break
    if isGray == False:
        break

# Convert the image to grayscale mode
if not isGray:
    gray = Image.new("L", (width, height))
    for i in range(width):
        for j in range(height):
            red, green, blue = imgNew.getpixel((i, j))
            newPxl = (red + green + blue) / 3
            newPxl = int(newPxl)
            gray.putpixel((i, j), newPxl)
    image = gray
    pixelsData_array = image.load()

#Threshold of image
for i in range(width):
    for j in range(height):
        if img.getpixel((i, j)) > int(thdKey):
            img.putpixel((i, j), 255)
        else:
            img.putpixel((i, j), 0)

plt.imshow(img, cmap = 'gray')
plt.title("treshold key is 110")
```
